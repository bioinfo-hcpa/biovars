```python
# Translate abraom annotations to gnomad

annotation_translation = {
    "nonsynonymous SNV": "missense_variant",
    "synonymous SNV": "synonymous_variant",
    "nonframeshift insertion": "inframe_insertion",
    "frameshift deletion": "frameshift_variant",
    "nonframeshift deletion": "inframe_deletion",
    "intronic": "intron_variant",
    "frameshift insertion": "frameshift_variant",
    "stopgain": "stop_gained",
    "UTR3": "3_prime_UTR_variant",
    "UTR5": "5_prime_UTR_variant"
}

a38_translated_ann = []
for ann in a38['Annotation'].values:
    if ann in annotation_translation:
        a38_translated_ann.append(annotation_translation[ann])
    else:
        a38_translated_ann.append(ann)
a38['Annotation'] = a38_translated_ann
```

```python
# Concatenate different variants

g38_no_overlap = g38.loc[~g38.index.isin(same_indexes)]
a38_no_overlap = a38.loc[~a38.index.isin(same_indexes)]

diff = pd.concat([g38_no_overlap, a38_no_overlap])
diff = diff.fillna(value=0)
```

```python
# Concatenate overlapped variants

a38_rep = a38.loc[same_indexes]
g38_rep = deepcopy(g38.loc[same_indexes])

new_allele_freq = []
for row in g38_rep.iterrows():
    variant_id = row[0]
    content = row[1]
    
    g38_rep.at[variant_id, 'Allele Count'] +=  a38_rep.loc[variant_id, 'Allele Count']
    g38_rep.at[variant_id, 'Allele Number'] +=  a38_rep.loc[variant_id, 'Allele Number']
    g38_rep.at[variant_id, 'Number of Homozygotes'] +=  a38_rep.loc[variant_id, 'Number of Homozygotes']
    g38_rep.at[variant_id, 'Number of Hemizygotes'] +=  a38_rep.loc[variant_id, 'Number of Hemizygotes']
    
    if g38_rep.at[variant_id, 'Allele Number'] != 0:
        new_allele_freq.append(
            g38_rep.at[variant_id, 'Allele Count'] / g38_rep.at[variant_id, 'Allele Number']
        )
    else:
        new_allele_freq.append(0)

g38_rep['Allele Frequency'] = new_allele_freq
g38_rep = g38_rep.fillna(0)
g38_rep['Brazilian'] = a38_rep['Brazilian']
```

```python
version = "hg38"
variant_id_freqs = ls  # Just to run here in the notebook


def integrate_gnomad_abraom(gnomad_df, abraom_df):
    
    # Data processing
    gnomad_df = prepare_gnomad_integration(gnomad_df) 
    abraom_df = prepare_abraom_integration(abraom_df)
    
    # Common variants
    same_indexes = set(gnomad_df.index).intersection(set(abraom_df.index))
    
    # Concatenate uncommon
    gnomad_no_overlap = gnomad_df.loc[~gnomad_df.index.isin(same_indexes)]
    abraom_no_overlap = abraom_df.loc[~abraom_df.index.isin(same_indexes)]

    uncommon = pd.concat([gnomad_no_overlap, abraom_no_overlap])
    uncommon = uncommon.fillna(value=0)
    
    # Concatenate common
    common = gnomad_abraom_concatenate_common_variants(gnomad_df, abraom_df, same_indexes)
    
    # Concatenate everything ordering by Chromosome and Location
    return pd.concat([uncommon, common]).sort_values(['Chromosome', 'Location'], ascending=[True, True])
    


def prepare_gnomad_integration(gnomad_df):
    processed_df = gnomad_df.drop('Flags', 1)
    processed_df = processed_df.set_index('Variant ID')
    return processed_df


def prepare_abraom_integration(abraom_df):
    processed_df = abraom_df.rename(columns={'Position': 'Location'})
    if 'GATK Filter' in processed_df.columns:
        processed_df = processed_df.drop('GATK Filter', 1)
    
    #variant_id_freqs = pyabraom.Variant_ID_biovar(abraom_df, version)
    
    processed_df['Variant ID'] = np.array(variant_id_freqs)[:,0]
    processed_df['Brazilian ABraOM'] = np.array(variant_id_freqs)[:,1]
        
    processed_df = processed_df.set_index('Variant ID')
    processed_df['Annotation'] = translate_abraom_annotations(processed_df['Annotation'].values)
    return processed_df
    
    

def translate_abraom_annotations(annoation_col):
    
    annotation_translation = {
        "nonsynonymous SNV": "missense_variant",
        "synonymous SNV": "synonymous_variant",
        "nonframeshift insertion": "inframe_insertion",
        "frameshift deletion": "frameshift_variant",
        "nonframeshift deletion": "inframe_deletion",
        "intronic": "intron_variant",
        "frameshift insertion": "frameshift_variant",
        "stopgain": "stop_gained",
        "UTR3": "3_prime_UTR_variant",
        "UTR5": "5_prime_UTR_variant"
    }

    translated_col = []
    for ann in annoation_col:
        if ann in annotation_translation:
            translated_col.append(annotation_translation[ann])
        else:
            translated_col.append(ann)
            
    return translated_col



def gnomad_abraom_concatenate_common_variants(gnomad_df, abraom_df, same_indexes):
    abraom_rep = abraom_df.loc[same_indexes]
    gnomad_rep = deepcopy(gnomad_df.loc[same_indexes])
    gnomad_rep['Brazilian ABraOM'] = abraom_rep['Brazilian ABraOM']

    new_allele_freq = []
    for row in gnomad_rep.iterrows():
        variant_id = row[0]
        content = row[1]

        gnomad_rep.at[variant_id, 'Allele Count'] +=  abraom_rep.loc[variant_id, 'Allele Count']
        gnomad_rep.at[variant_id, 'Allele Number'] +=  abraom_rep.loc[variant_id, 'Allele Number']
        gnomad_rep.at[variant_id, 'Number of Homozygotes'] +=  abraom_rep.loc[variant_id, 'Number of Homozygotes']
        gnomad_rep.at[variant_id, 'Number of Hemizygotes'] +=  abraom_rep.loc[variant_id, 'Number of Hemizygotes']

        if gnomad_rep.at[variant_id, 'Allele Number'] != 0:
            new_allele_freq.append(
                gnomad_rep.at[variant_id, 'Allele Count'] / gnomad_rep.at[variant_id, 'Allele Number']
            )
        else:
            new_allele_freq.append(0)

    gnomad_rep['Allele Frequency'] = new_allele_freq
    gnomad_rep = gnomad_rep.fillna(0)
    return gnomad_rep
```
